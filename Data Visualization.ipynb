# Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
# x axis values
x = [1,2,3,4,5]

# y axis values
y = [3,2,6,4,7]
```

```python
# Naming the labels
plt.xlabel('Sales')
plt.ylabel('Profit')

# title of graph
plt.title('Graph!!!')

# plotting the points
plt.plot(x,y)
```

```python
a = [12,72,23,46,54]
b = [30,24,46,74,70]

# Naming the labels
plt.xlabel('Day')
plt.ylabel('temp')

# title of graph
plt.title('temp!!!')

# plotting the points
plt.plot(a, b)
```

```python
a = [12,72,23,46,54]
b = [30,24,46,74,70]

# Naming the labels
plt.xlabel('Day')
plt.ylabel('temp')

# title of graph
plt.title('temp!!!')

# plotting the points
plt.plot(a, b, linestyle='dotted')
```

```python
a = [12,72,23,46,54]
b = [30,24,46,74,70]

# Naming the labels
plt.xlabel('Day')
plt.ylabel('temp')

# title of graph
plt.title('temp!!!')

# plotting the points
plt.plot(a, b, color='limegreen')
```

```python
x = [1,2,3,4,5]
y = [3,2,6,4,7]

plt.bar(x,y, color='lightgreen')
plt.plot(x,y, color='black')
```

```python
x = [1,2,3,4,5]
y = [3,2,6,4,7]

plt.scatter(x,y)
```

```python
# Subplots

# plot 1
x = np.array([1,2,3,4,5])
y = np.array([2,6,23,56,89])

# plot 2
a = np.array([2,3,5,6,7])
b = np.array([14,23,45,12,36])

# # plot 3
# c = random.randint(5,size=(10))
# d = random.randint(100,size=(10))

plt.subplot(2,1,1)
plt.plot(x,y)

plt.subplot(2,1,2)
plt.plot(a,b)
```

```python

# plot 1
x1 = random.randint(5,size=(5))
y1 = random.randint(50,size=(5))

# plot 2
x2 = random.randint(5,size=(5))
y2 = random.randint(50,size=(5))

# plot 3
x3 = random.randint(5,size=(5))
y3 = random.randint(50,size=(5))

# plot 4
x4 = random.randint(5,size=(5))
y4 = random.randint(50,size=(5))


plt.subplot(2,2,1)
plt.plot(x1,y1)

plt.subplot(2,2,2)
plt.plot(x2,y2)

plt.subplot(2,2,3)
plt.plot(x3,y3)

plt.subplot(2,2,4)
plt.plot(x4,y4)
```

```python
data = random.normal(170,10,250)

# the hist() function will read the array and produce a histogram
# set the bins value 30

plt.hist(data,bins=10)
plt.show()
```

```python
cars = ['a','b','c','d','e']

data = [51,41,81,64,61]

plt.pie(data, labels=cars)

plt.show()
```
