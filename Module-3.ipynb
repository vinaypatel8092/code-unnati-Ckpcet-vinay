```python
# Q-1.Create Python Tkinter GUI program to add a button in your application using tkinter module.

import tkinter as tk

# Create a function to be called when the button is clicked
def button_clicked():
    print("Button clicked")

# Create a Tkinter window
window = tk.Tk()
window.title("My GUI")

# Create a button and add it to the window
button = tk.Button(window, text="Click me!", height=1, width=35, command=button_clicked)
button.pack()

# Run the Tkinter event loop
window.mainloop()
```

```python
# Q-2.Create Python Jupyter notebook of Sk-Learn program using Linear Regression(LR) to classify cancer samples diagnosed as Benign or Malignant 
# Dataset:
# https://raw.githubusercontent.com/AgustinCdeCelis/Cancer-Logistic-Regression/main/data.csv


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

cancer = pd.read_csv('https://raw.githubusercontent.com/AgustinCdeCelis/Cancer-Logistic-Regression/main/data.csv')
```

```python
cancer
```

```python
cancer.info()
```

```python
cancer.columns
```

```python
X = cancer[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
y = cancer['diagnosis']
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y)
y = le.transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state = 101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_percentage_error(y_test,predictions)
```

```python
mean_squared_error(y_test,predictions)
```

```python
# Q-3.Create Python Jupyter notebook of Sk-Learn program using Support Vector Machine(SVM) to classify cancer samples diagnosed as Benign or Malignant 
# Dataset: https://raw.githubusercontent.com/AgustinCdeCelis/Cancer-Logistic-Regression/main/data.csv
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
```

```python
cancer = pd.read_csv('https://raw.githubusercontent.com/AgustinCdeCelis/Cancer-Logistic-Regression/main/data.csv')
```

```python
cancer
```

```python
y = cancer['diagnosis']
```

```python
x = cancer[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, random_state = 4)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Generating Model

clf = svm.SVC(kernel='rbf') # Linear Kernel
print(y_train.shape)
```

```python
# Train the model using the training sets

clf.fit(x_train, y_train) 
```

```python
# Predict the response for the test dataset

y_pred = clf.predict(x_test) 
```

```python
y_test
```

```python
y_pred
```

```python
# Evaluation

metrics.accuracy_score(y_test, y_pred)
```
